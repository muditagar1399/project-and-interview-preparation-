```python
import requests
import pyodbc
from datetime import datetime

# SQL Server Connection
conn = pyodbc.connect(
    r"Driver={SQL Server};"
    r"Server=localhost\SQLEXPRESS;"
    r"Database=crytpo;"
    r"Trusted_Connection=yes;"
)

cursor = conn.cursor()

# CoinGecko API
url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 300,
    "page": 1,
    "sparkline": "false"
}

response = requests.get(url, params=params)
coins = response.json()

# Insert data into SQL Server
for coin in coins:

    cursor.execute("""
    INSERT INTO crytpo
    (
        coinName,
        symbol,
        current_price,
        marketcap,
        volumne24h,
        change24h,
        marketrank,
        lastupdated
    )
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
    """,
    coin["name"],
    coin["symbol"],
    coin["current_price"],
    coin["market_cap"],
    coin["total_volume"],
    coin["price_change_percentage_24h"],
    coin["market_cap_rank"],
    datetime.now()
    )

conn.commit()

print(f"{len(coins)} records inserted successfully!")

cursor.close()
conn.close()
```
